```python
#import data of Orca
import pandas as pd
file ='Number of Southern Resident killer whales (3).csv'
profile_read = pd.read_csv(file,delimiter = ',')
display(profile_read)
```

```python
# ploting figure of total number
import matplotlib.pyplot as plt
fig, axs = plt.subplots(figsize = (10,5))
x = profile_read['Census year']
y = profile_read['All pods']
axs.plot(x, y, marker= 'o', label='All pods')
axs.grid()
axs.legend()
axs.set_title('Total Southern Resident Killer Whales Population', fontsize=15)
axs.set_xlabel('Year', fontsize=15)
axs.set_ylabel('Number of Whales', fontsize=15)
#ploting figure of three different pods
fig, axs = plt.subplots(figsize = (10,5))
x = profile_read['Census year']
y1 = profile_read['L pod']
y2= profile_read['J pod']
y3= profile_read['K pod']
axs.plot(x, y1, marker= 'o', label='L pod')
axs.plot(x, y2, marker= 'o', label='J pod')
axs.plot(x, y3, marker= 'o', label='K pod')
axs.legend()
axs.grid()
axs.set_title('Southern Resident Killer Whales Population', fontsize=15)
axs.set_xlabel('Year', fontsize=15)
axs.set_ylabel('Number of Whales', fontsize=15)
```

```python
#import data of Chinook
import pandas as pd
file1 ='salmon population3.csv'
profile_read1 = pd.read_csv(file1,delimiter = ',')
display(profile_read1)
```

```python
# ploting figure of chinook
import matplotlib.pyplot as plt
fig, axs = plt.subplots(figsize = (10,5))

x = profile_read1['Year']
y = profile_read1['Spring']
y2= profile_read1['Fall']
axs.plot(x,y,marker='o',label='Spring')
axs.plot(x,y2,marker='o',label='Fall')

axs.legend()
axs.grid()
axs.set_title('CENTRAL VALLEY Chinook Population', fontsize=15)
axs.set_xlabel('Year',fontsize=15)
axs.set_ylabel('Number of Chinook',fontsize=15)
```
